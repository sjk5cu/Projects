# Querying

Who are the highest rated players?

```python
myquery = ''' 

SELECT player_name, AVG(overall_rating) as mean_rating
FROM Player_Attributes
INNER JOIN Player p
    ON p.player_api_id = Player_Attributes.player_api_id
GROUP BY player_name
ORDER BY mean_rating DESC

''' 

pd.read_sql_query(myquery, con=con)[0:10]
```

What is the win/loss record for each team over the 8 seasons?

```python
matches = pd.read_sql_query("SELECT * FROM Match", con)
```

```python
matches.head()
```

```python
myquery2 = ''' 

SELECT home_team_api_id, team_long_name, SUM(home_team_goal>away_team_goal) AS win, SUM(home_team_goal < away_team_goal) AS loss, SUM(home_team_goal = away_team_goal) AS draw
FROM Match
INNER JOIN Team t
    ON t.team_api_id = Match.home_team_api_id
GROUP BY team_long_name
ORDER BY win DESC
''' 

pd.read_sql_query(myquery2, con=con)
```

Breaking up the teams into tiers

```python
myquery3 = ''' 

SELECT home_team_api_id, team_long_name, win, loss, draw, CASE
    WHEN win <27 then 'tier 5'
    WHEN win BETWEEN 27 AND 53 THEN 'tier 4'
    WHEN win BETWEEN 53 AND 79 THEN 'tier 3'
    WHEN win BETWEEN 79 AND 105 THEN 'tier 2'
    WHEN win BETWEEN 105 AND 132 THEN 'tier 1'
    ELSE NULL
    END AS tier
FROM(
SELECT home_team_api_id, team_long_name, SUM(home_team_goal>away_team_goal) AS win, SUM(home_team_goal < away_team_goal) AS loss, SUM(home_team_goal = away_team_goal) AS draw
FROM Match
INNER JOIN Team t
    ON t.team_api_id = Match.home_team_api_id
GROUP BY team_long_name
ORDER BY win DESC
)
''' 

pd.read_sql_query(myquery3, con=con)
```

Finding play characteristics for each team

```python
myquery4 = ''' 

SELECT home_team_api_id, AVG(buildUpPlaySpeed),AVG(buildUpPlayDribbling), AVG(buildUpPlayPassing), AVG(chanceCreationPassing), AVG(chanceCreationCrossing), AVG(chanceCreationShooting), AVG(defencePressure), AVG(defenceAggression)
FROM Match
INNER JOIN Team t
    ON t.team_api_id = Match.home_team_api_id
INNER JOIN Team_Attributes ta
    ON ta.team_api_id = Match.home_team_api_id
GROUP BY home_team_api_id


''' 

pd.read_sql_query(myquery4, con=con)
```

Joining two subqueries

```python
myquery5 = ''' 
SELECT *
FROM(
    SELECT home_team_api_id, team_long_name, win, loss, draw, CASE
    WHEN win <27 then 'tier 5'
    WHEN win BETWEEN 27 AND 53 THEN 'tier 4'
    WHEN win BETWEEN 53 AND 79 THEN 'tier 3'
    WHEN win BETWEEN 79 AND 105 THEN 'tier 2'
    WHEN win BETWEEN 105 AND 132 THEN 'tier 1'
    ELSE NULL
    END AS tier
    FROM(
        SELECT home_team_api_id, team_long_name, SUM(home_team_goal>away_team_goal) AS win, SUM(home_team_goal < away_team_goal) AS loss, SUM(home_team_goal = away_team_goal) AS draw
        FROM Match
        INNER JOIN Team t
            ON t.team_api_id = Match.home_team_api_id
        GROUP BY team_long_name
        ORDER BY win DESC
        )
    ) y   
JOIN(
    SELECT home_team_api_id, AVG(buildUpPlaySpeed),AVG(buildUpPlayDribbling), AVG(buildUpPlayPassing), AVG(chanceCreationPassing), AVG(chanceCreationCrossing), AVG(chanceCreationShooting), AVG(defencePressure), AVG(defenceAggression)
    FROM Match
    INNER JOIN Team t
        ON t.team_api_id = Match.home_team_api_id
    INNER JOIN Team_Attributes ta
        ON ta.team_api_id = Match.home_team_api_id
    GROUP BY home_team_api_id 
    ) z
    ON z.home_team_api_id = y.home_team_api_id


''' 

pd.read_sql_query(myquery5, con=con)
```
